## Performing Basic SQL Queries

```
%sql
-- Finding the highest number of units for a particular category
select count(t.units)as units, p.category
from transaction_table t
join product_table p
on t.upc = p.upc
group by category
order by units desc
```

```
%sql
-- Finding the average money spent for the store_name
select avg(t.spend)avg_spend, s.store_name
from transaction_table t
join store_table s
on t.store_num = s.store_id
group by store_name
order by avg_spend desc
```

```
%sql
-- Finding the average baskets for each store
select avg(avg_weekly_baskets) as avg_baskets, store_name
from store_table
group by store_name
order by avg_baskets desc
```

```
%sql
-- Finding total number of baskets for each city
select sum(avg_weekly_baskets) as total_baskets, city
from store_table
group by city
order by total_baskets desc
```

```
%sql
select s.store_name, sum(t.feature) as total_features--, sum(t.display) as totaL_displays, sum(t.tpr_only) as sum_tpr
from transaction_table t
join store_table s
on t.store_num = s.store_id
group by store_name
order by total_features desc
```

```
%sql
select s.store_name, sum(t.display) as totaL_displays--, sum(t.tpr_only) as sum_tpr
from transaction_table t
join store_table s
on t.store_num = s.store_id
group by store_name
order by total_displays desc
```

```
%sql
select s.store_name, sum(t.tpr_only) as sum_tpr, sum(t.hhs) as hs
from transaction_table t
join store_table s
on t.store_num = s.store_id
group by store_name
order by sum_tpr desc
```

## Create Models for Prediction

```
# Import Libraries
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.regression import LinearRegression
from pyspark.ml import Pipeline
```

### Predicting Category

#### Model Preprocessing

##### Random Forest

```
# Selecting data for linear model
data = products_transactions_store.select(['category', 'product_size', 'price', 'units'])

# Drop NAs
print('Length Before Dropping NAs: ', data.count())
data = data.dropna()
print('Length After Dropping NAs: ', data.count())
```

```
display(data)
```

```
# Split the data into train and test
train, test = data.randomSplit([0.7, 0.3])
```

```
# Create a string indexer for the category column
category_indexer = StringIndexer(inputCol='category', outputCol='category_index', handleInvalid='keep')
product_size_indexer = StringIndexer(inputCol='product_size',outputCol='product_size_index', handleInvalid='keep')
```

```
# Create a vector assembler for the features
assembler = VectorAssembler(inputCols=['price', 'product_size_index', 'units'], 
                            outputCol='features')
```

```
# Create a random forest classifier
rf = RandomForestClassifier(labelCol='category_index', featuresCol='features')

# Create a pipeline
pipeline = Pipeline(stages=[category_indexer, product_size_indexer, assembler, rf])
```

#### Fit, Transform and Predict

```
# Fit the pipeline to the training data
model = pipeline.fit(train)
```

```
# Transform the test data
results = model.transform(test)
```

```
# See the results
results.select('category', 'category_index', 'prediction').show(20)
```

#### Model Evaluation

```
# Import evaluation library
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from sklearn.metrics import confusion_matrix
```

```
# confusion matrix
results.groupBy('category_index', 'prediction').count().show()
```

```
# plot confusion matrix
results.groupBy('category_index', 'prediction').count().toPandas().plot(kind='bar', x='category_index', y='count', color = '#065842', figsize = (25,10))
```

```
# True Positive, Negative, False Positive, Negative
tp = results.filter('category_index == 0 AND prediction == 0').count()
tn = results.filter('category_index == 1 AND prediction == 1').count()
fp = results.filter('category_index == 1 AND prediction == 0').count()
fn = results.filter('category_index == 0 AND prediction == 1').count()

# print confusion matrix
print('True Positive: {}'.format(tp))
print('True Negative: {}'.format(tn))
print('False Positive: {}'.format(fp))
print('False Negative: {}'.format(fn))
```

```
# recall, precision, f1 score, accuracy
evaluator = MulticlassClassificationEvaluator(labelCol='category_index', predictionCol='prediction', metricName='accuracy')
recall = tp / (tp + fn)
precision = tp / (tp + fp)
f1 = 2 * (precision * recall) / (precision + recall)
accuracy = evaluator.evaluate(results)

# print recall, precision, f1 score
print('Recall: {}'.format(recall))
print('Precision: {}'.format(precision))
print('F1 Score: {}'.format(f1))
print('Accuracy: {}'.format(accuracy))
```

### Predicting Units Sold

#### Model Preprocessing

##### Linear Regression

```
# Selecting data for linear model
data = products_transactions_store.select(['description','manufacturer','category','sub_category','product_size','units','visits','hhs','spend','price','base_price'    ,'feature','display','tpr_only','store_name','city','state','msa','segment','size','avg_weekly_baskets'])

# Drop NAs
print('Length Before Dropping NAs: ', data.count())
data = data.dropna()
print('Length After Dropping NAs: ', data.count())
```

```
display(data)
```

```
#Spliting data
train_data,test_data = data.randomSplit([0.7,0.3])
```

```
# Use StringIndexer to convert the categorical columns to hold numerical data
 
description_indexer = StringIndexer(inputCol='description',outputCol='description_index',handleInvalid='keep')
manufacturer_indexer = StringIndexer(inputCol='manufacturer',outputCol='manufacturer_index',handleInvalid='keep')
category_indexer = StringIndexer(inputCol='category',outputCol='category_index',handleInvalid='keep')
sub_category_indexer = StringIndexer(inputCol='sub_category',outputCol='sub_category_index',handleInvalid='keep')
product_size_indexer = StringIndexer(inputCol='product_size',outputCol='product_size_index',handleInvalid='keep')
store_name_indexer = StringIndexer(inputCol='store_name',outputCol='store_name_index',handleInvalid='keep')
city_indexer = StringIndexer(inputCol='city',outputCol='city_index',handleInvalid='keep')
state_indexer = StringIndexer(inputCol='state',outputCol='state_index',handleInvalid='keep')
segment_indexer = StringIndexer(inputCol='segment',outputCol='segment_index',handleInvalid='keep')
```

```
# Vector assembler is used to create a vector of input features
 
assembler = VectorAssembler(inputCols=['description_index','manufacturer_index','category_index',
                                       'sub_category_index','product_size_index','store_name_index',
                                       'city_index','state_index','segment_index'],
                            outputCol="features")
```

```
# Pipelining
pipe = Pipeline(stages=[description_indexer,manufacturer_indexer,category_indexer,sub_category_indexer,
                        product_size_indexer,store_name_indexer,city_indexer,
                        state_indexer,segment_indexer,assembler])
```

#### Fit, Transform, Predict

```
# Fitting model on train data
fitted_pipe = pipe.fit(train_data)

# Transforming data with model
train_data = fitted_pipe.transform(train_data)
```

```
# Defining the linear regression model
lr_model = LinearRegression(labelCol='units')

# Fitting model
fit_model = lr_model.fit(train_data.select(['features','units']))
```

```
# Run model on test data set
test_data=fitted_pipe.transform(test_data)

# Make prediction
results = fit_model.transform(test_data)
```

```
# Show predicted results
results.select(['units','prediction']).show()
```

#### Model Evaluation

```
# Printing coefficients for each column
coeffiecients = zip(data.columns[:-1], fit_model.coefficients)
for coeff in coeffiecients:
    print(coeff)
```

```
# Create evaluation
test_results = fit_model.evaluate(test_data)

# See residuals
test_results.residuals.show()
```

```
# Summarize the model over the training set and print out some metrics
trainingSummary = fit_model.summary

# Print Scores
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("r2: %f" % trainingSummary.r2)
print("Degrees of Freedom: %f " % trainingSummary.degreesOfFreedom)

# Print P-Values for each column
print("\nP-Values")
p_values = zip(data.columns[:-1], trainingSummary.pValues)
for pval in p_values:
    print(pval)
```

#### Attempting a different model (K-Means Cluster)

```
# Import necessary libraries and define the model
from pyspark.ml.clustering import KMeans

kmeans_model = KMeans(k=4)
```

```
# Fit and Transform the model
final_data = pipe.fit(data).transform(data)

fit_model_kmeans = kmeans_model.fit(final_data)
```

```
# Calculate the Within Set Sum of Squared Error

wssse = fit_model_kmeans.summary.trainingCost
print("The within set sum of squared error of the mode is {}".format(wssse))
```

```
# Get the centroids of the clusters
centers = fit_model_kmeans.clusterCenters()

print("Cluster Centers")
index=1
for cluster in centers:
    print("Centroid {}: {}".format(index,cluster))
    print("")
    index+=1
```

```
# Make prediction
results = fit_model_kmeans.transform(final_data)

results.select(['description_index','manufacturer_index','category_index',
                                       'sub_category_index','product_size_index','store_name_index',
                                       'city_index','state_index','segment_index','prediction']).show()
```

```
# Check the majority number of predicted clusters
results.groupby('prediction').count().sort('count', ascending = False).show()
```

### Predict Size of Store based on Basket Sales

##### Linear Regression

```
# Selecting the dependent and the independent variables that are identified as most useful attributes to estimate duration
 
data = products_transactions_store.select(['size','avg_weekly_baskets'])

# Drop NAs
print('Length Before Dropping NAs: ', data.count())
data = data.dropna()
print('Length After Dropping NAs: ', data.count())
```

```
# Create a 70-30 train test split
 
train_data,test_data=data.randomSplit([0.7,0.3])
```

```
# Vector assembler is used to create a vector of input features
assembler = VectorAssembler(inputCols=['size'],
                            outputCol="features")

# Pipeline is used to pass the data through indexer and assembler simultaneously. Also, it helps to pre-rocess the test data
# in the same way as that of the train data
pipe = Pipeline(stages=[assembler])
```

#### Fit, Transform, Predict

```
# Fit and Transform
fitted_pipe=pipe.fit(train_data)

train_data=fitted_pipe.transform(train_data)
```

```
# Create an object for the Linear Regression model
lr_model = LinearRegression(labelCol='avg_weekly_baskets')

# Fit the model on the train data
fit_model = lr_model.fit(train_data.select(['features','avg_weekly_baskets']))
```

```
# Transform the test data using the model to predict the duration
test_data=fitted_pipe.transform(test_data)

# Store the results in a dataframe
results = fit_model.transform(test_data)
display(results)
```

```
# View prediction with actual side-by-side
results.select(['avg_weekly_baskets','prediction']).show()
```

#### Evaluate the model

```
# Printing coefficients for each column
coeffiecients = zip(data.columns[:-1], fit_model.coefficients)
for coeff in coeffiecients:
    print(coeff)
```

```
# Store evaluations
test_results = fit_model.evaluate(test_data)

# Check residuals
test_results.residuals.show()
```

```
# Summarize the model over the training set and print out some metrics
trainingSummary = fit_model.summary

# Print Scores
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("r2: %f" % trainingSummary.r2)
print("Degrees of Freedom: %f " % trainingSummary.degreesOfFreedom)

# Print P-Values for each column
print("\nP-Values")
p_values = zip(data.columns[:-1], trainingSummary.pValues)
for pval in p_values:
    print(pval)
```

### Predicting the effect of State based on Basket Sales

##### Decision Tree

```
# Import library
from pyspark.ml.classification import DecisionTreeClassifier
```

```
# Selecting the dependent and the independent variables that are identified as most useful attributes to estimate duration
 
data = store_df.select(['state','avg_weekly_baskets'])

# Drop NAs
print('Length Before Dropping NAs: ', data.count())
data = data.dropna()
print('Length After Dropping NAs: ', data.count())
```

```
# Import Library
from pyspark.sql.functions import when

# Convert each state to a numeric dummy value
df_pred = data.withColumn("state", when(data.state == "IN","1"). \
      when(data.state == "KY","2"). \
      when(data.state == "OH","3"). \
      when(data.state == "TX","4") \
      .otherwise(data.state ))

# Verify
display(df_pred)
```

```
# Create an object for the Logistic Regression model
# Use the parameter maxBins and assign a value that is equal to or more than the number of categories in any sigle feature
 
dt_model = DecisionTreeClassifier(labelCol='state',maxBins=5000)
```

```
# Vector assembler is used to create a vector of input features
 
assembler = VectorAssembler(inputCols=["avg_weekly_baskets"],
                            outputCol="features")
```

```
# Pipeline is used to pass the data through indexer and assembler simultaneously. Also, it helps to pre-rocess the test data
# in the same way as that of the train data. It also 
 
pipe = Pipeline(stages=[assembler,dt_model])
```

```
# Convert values to float
df_pred = df_pred.withColumn("avg_weekly_baskets", df_pred["avg_weekly_baskets"].cast('float'))
df_pred = df_pred.withColumn("state", df_pred["state"].cast('float'))
```

```
# Split into train and test
train_data,test_data=df_pred.randomSplit([0.7,0.3])
```

```
# Fit and Transform
fit_model=pipe.fit(train_data)

# Predict
results = fit_model.transform(test_data)
```

```
results.select(['state','prediction']).show()
```

#### Evaluate the model

```
ACC_evaluator = MulticlassClassificationEvaluator(
    labelCol="state", predictionCol="prediction", metricName="accuracy")

accuracy = ACC_evaluator.evaluate(results)

print("The accuracy of the model is {}".format(accuracy))
```
